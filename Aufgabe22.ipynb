```python
import maabara as ma
import numpy as np
import scipy as sp
import sympy as sym
import matplotlib as mpl 
import matplotlib.pyplot as plt
import math
import matplotlib.gridspec as gridspec
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from mpl_toolkits.mplot3d.axes3d import Axes3D
mpl.rcParams['text.usetex']=True
mpl.rcParams['text.latex.unicode']=True
mpl.rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})

#Gitterbeugung:
N=4.
d=3000.0e-9
L=1500e-9
ll=632.816e-9 #Lambda
def I(x,l):
    return 1/N**2*(np.sin(np.pi*l/ll*x)**2)/(np.pi*l/ll*x)**2*(np.sin((N*np.pi*d)/(ll)*x)**2)/(np.sin(np.pi*d/ll*x))**2
def einh(x,l):
    return (np.sin(np.pi*l/ll*x)**2)/(np.pi*l/ll*x)**2

plt.clf()
plt.figure()
ax=plt.subplot(111)
x=np.linspace(-4*ll/d,4*ll/d,100000)  #Definitionsbereich
a=(1.5**2-(2*1.5+1)*1+(1.5+2)/1.5)/(32.*(1.5-1.)**2)
ax.plot(x,I(x,L))
ax.plot(x,einh(x,L))
ax.set_title('Aufgabe 22: Beugung am Gitter') #Titel
ax.legend(numpoints=1, markerscale=1.5, fontsize=10,ncol=3,fancybox=True, shadow=True, loc=0)
ax.set_ylabel(r'Verhältnis der Intensität zur maximalen Intensität $\frac{I(\theta)}{I_0}$')
ax.set_xlabel(r'$\sin\theta$')
#ax.set_xticks(np.arange(-1, 1.2, 0.2))
ax.grid(True)

plt.savefig("plot1.pdf",transparent=True,format="pdf",bbox_inches='tight')

#3D PLOT
plt.clf()
fig = plt.figure()
ax = plt.axes(projection='3d')

x = np.linspace(4*ll/d,-4*ll/d,10000)
y = np.linspace(3000e-9,500e-9,400)
X, Y = np.meshgrid(x, y)
Z = I(X, Y)
surf=ax.plot_surface(X, Y, Z,cmap=mpl.cm.RdBu,linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_title('Abhängigkeit des Beugungsbildes von der Spaltbreite $L$',fontsize=20) #Titel
ax.set_ylabel(r'Spaltbreite $L$ [m]',fontsize=20)
ax.set_xlabel(r'$\sin\theta$',fontsize=20)
ax.set_zlabel(r'Verhältnis der Intensität zur maximalen Intensität $\frac{I(\theta)}{I_0}$',fontsize=20)
plt.ticklabel_format(style='sci', scilimits=(-3,4))
plt.show()



plt.clf()
plt.figure()
plt.ticklabel_format(style='sci', scilimits=(0,0))
ax2=plt.subplot(111)
x=np.linspace(0,2,1000)
b=((4*(800e-9)**3*0.1)/(np.pi)**3)**(1/4.)
n=1.5
ax2.plot(x,b*((n**2-(2*n+1)*x+(n+2)/n*x**2)/(32.*(n-1.)**2))**(1/4.),'b-',label='$n=1,5$')
n=1.7
ax2.plot(x,b*((n**2-(2*n+1)*x+(n+2)/n*x**2)/(32.*(n-1.)**2))**(1/4.),'g-',label='$n=1,7$')
n=1.9
ax2.plot(x,b*((n**2-(2*n+1)*x+(n+2)/n*x**2)/(32.*(n-1.)**2))**(1/4.),'r-',label='$n=1,9$')

ax2.set_xlabel('Formfaktor $K$')
ax2.set_ylabel('minimaler Radius $w_0\'$ [m]')
ax2.grid(True)
ax2.legend(loc=2)
ax2.set_xticks(np.arange(0, 2, 0.1))
plt.savefig("plot2.pdf",transparent=True,format="pdf",bbox_inches='tight')
```
